```python
import keras
from keras.models import Sequential
from keras.layers import Dense, Conv2D , MaxPooling2D , Flatten , Dropout , BatchNormalization
from keras.preprocessing.image import ImageDataGenerator
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report,confusion_matrix
from keras.callbacks import ReduceLROnPlateau
from keras_preprocessing.image import ImageDataGenerator
from keras import regularizers, optimizers
```

```python
from keras.models import load_model
model = load_model('/kaggle/input/notebook6817bfa5c9/best_model.h5')
```

```python
test_df = pd.read_csv("/kaggle/input/early-detection-of-3d-printing-issues/test.csv")
```

```python
datagen=ImageDataGenerator(rescale=1./255.)

test_generator=datagen.flow_from_dataframe(
dataframe=test_df,
directory="/kaggle/input/early-detection-of-3d-printing-issues/images/",
x_col="img_path",
y_col = None,
batch_size=16,
seed=42,
shuffle=False,
class_mode=None,
target_size=(256,256))
```

```python
y_pred = model.predict(test_generator)
```

```python
y_pred
```

```python
y_classes = []
for x in y_pred:
    if x >= 0.5:
        y_classes.append(1)
    else:
        y_classes.append(0)
print(y_classes)
```

```python
np.unique(y_classes, return_counts=True)
```

```python
test_df = test_df.drop(columns=['printer_id', 'print_id'])
```

```python
test_df['has_under_extrusion'] = y_classes
```

```python
test_df.to_csv("submission.csv", index = False)
test_df.head()
```
